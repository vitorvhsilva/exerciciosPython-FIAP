```python
#1 
valor_valido = False

while not valor_valido:
    num = input('Digite um número entre 0 e 10: ')
    if num.replace('-','').isdigit():
        num = int(num)
        if num < 0 or num > 10:
            print('Valor incorreto')
        else:
            print(f'Nota: {num}')
            valor_valido = True
    else:
        print('Formato incorreto')

print('Processo completo!')
```

```python
#2

from getpass import getpass

loop = True

while loop:
    user = input('Digite seu usuário: ')
    senha = getpass('Digite sua senha: ')
    if user == senha:
        print('Informações iguais, digite novamente')
    else:
        loop = False

print('Processo completo!')
```

```python
#3

loop = True
infos = []

while loop:
    user = input('Digite seu usuário: ')
    if len(user) <= 3:
        print('Informações inválidas, digite novamente')
    else:
        print('Informações corretas')
        infos.append(user)
        loop = False

loop = True

while loop:
    idade = input('Digite sua idade: ')
    idade = int(idade)
    if idade > 150 or idade < 0:
        print('Informações inválidas, digite novamente')
    else:
        print('Informações corretas')
        infos.append(idade)
        loop = False

loop = True

while loop:
    salario = input('Digite seu salario: ')
    salario = float(salario)
    if salario < 0:
        print('Informações inválidas, digite novamente')
    else:
        print('Informações corretas')
        infos.append(salario)
        loop = False

loop = True

while loop:
    sexo = input('Digite seu sexo: ')
    if sexo.lower() in ['f', 'm']:
        print('Informações corretas')
        infos.append(sexo.lower())
        loop = False
    else:
        print('Informações inválidas, digite novamente')

loop = True

while loop:
    estado = input('Digite seu estado civil: ')
    if estado.lower() in ['s', 'c', 'v', 'd']:
        print('Informações corretas')
        infos.append(estado.lower())
        loop = False
    else:
        print('Informações inválidas, digite novamente')

print(infos)
```

```python
#21

num = input('Digite um numero inteiro: ')
num = int(num)
primo = 0

for i in range(1, (num + 1)):
  if num % i == 0:
    primo += 1

if primo == 2:
  print(f'Numero {num} e primo')
else:
  print(f'Numero {num} nao e primo')
```

```python
#22

num = input('Digite um numero inteiro: ')
num = int(num)
divisiveis = []
primo = 0

for i in range(1, (num + 1)):
  if num % i == 0:
    primo += 1
    divisiveis.append(i)

if primo == 2:
  print(f'Numero {num} e primo')
else:
  print(f'Numero {num} nao e primo')
  print('Divisivel por: ', divisiveis)

```

```python
#23

num = input('Digite ate onde sera mostrados os primos: ')
num = int(num)
primos = []
nprimos = []

for x in range(1, (num + 1)):
  primo = 0
  for i in range(1, (x + 1)):
    if x % i == 0:
      primo += 1

  if primo == 2:
    primos.append(x)
  else:
    nprimos.append(x)

print(f'Numeros primos: {primos}')
print(f'Numeros nao primos: {nprimos}')
```
